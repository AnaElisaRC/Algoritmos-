```python
# Importa a estrutura de dados 'deque' da biblioteca 'collections'
from collections import deque

# Capacidades máximas das jarras
JARRA_1_CAPACID = 3   # jarra menor
JARRA_2_CAPACID = 7   # jarra maior

# Objetivo: obter exatamente 5 litros na jarra de 7 litros
OBJETIVO = 5

# Estado inicial: ambas as jarras estão vazias
estado_inicial = (0, 0)

# Função que gera todos os estados possíveis a partir de um estado atual
def obter_estados_seguintes(estado):
    x, y = estado  # x: quantidade na jarra 1, y: quantidade na jarra 2
    estados = []

    # 1. Encher completamente a jarra 1
    estados.append((JARRA_1_CAPACID, y))
    # 2. Encher completamente a jarra 2
    estados.append((x, JARRA_2_CAPACID))
    # 3. Esvaziar completamente a jarra 1
    estados.append((0, y))
    # 4. Esvaziar completamente a jarra 2
    estados.append((x, 0))
    # 5. Transferir da jarra 1 para a 2 até encher a 2 ou esvaziar a 1
    transferir = min(x, JARRA_2_CAPACID - y)
    estados.append((x - transferir, y + transferir))
    # 6. Transferir da jarra 2 para a 1 até encher a 1 ou esvaziar a 2
    transferir = min(y, JARRA_1_CAPACID - x)
    estados.append((x + transferir, y - transferir))

    return estados

# Função que implementa a busca em largura (BFS) para encontrar a solução
def bfs():
    fila = deque()       # fila para armazenar os estados a serem explorados
    visitado = set()     # conjunto para armazenar os estados já visitados
    predecessor = {}     # dicionário para rastrear o caminho de cada estado

    fila.append(estado_inicial)
    visitado.add(estado_inicial)
    predecessor[estado_inicial] = None

    while fila:
        atual = fila.popleft()
        x, y = atual

        # A linha abaixo foi comentada para limpar a saída.
        # Descomente se quiser ver todos os passos da busca.
        # print(f"Explorando -> Jarra {JARRA_1_CAPACID}L: {x}L, Jarra {JARRA_2_CAPACID}L: {y}L")

        # Verifica se atingimos o objetivo
        if y == OBJETIVO or x == OBJETIVO:  # Verifica se qualquer uma das jarras atingiu o objetivo
            print("\nSolução encontrada!")
            caminho = []
            # Reconstrói o caminho desde o estado inicial até o objetivo
            while atual:
                caminho.append(atual)
                atual = predecessor[atual]
            caminho.reverse()
            # Imprime a sequência de estados que leva à solução
            print("Passo a passo da solução:")
            for i, estado in enumerate(caminho):
                print(f"{i}: Jarra {JARRA_1_CAPACID}L: {estado[0]}L, Jarra {JARRA_2_CAPACID}L: {estado[1]}L")
            return

        # Gera e explora os próximos estados possíveis
        for estado_seguinte in obter_estados_seguintes(atual):
            if estado_seguinte not in visitado:
                visitado.add(estado_seguinte)
                predecessor[estado_seguinte] = atual
                fila.append(estado_seguinte)

    print("Nenhuma solução encontrada.")

# Executa o algoritmo de busca Breadth First Search
bfs()
```

```python
from collections import deque

def obter_estados_seguintes(estado, cap_j1, cap_j2):
    """Gera todos os estados possíveis a partir de um estado atual."""
    x, y = estado  # x: quantidade na jarra 1, y: quantidade na jarra 2
    estados = []

    # 1. Encher completamente a jarra 1
    estados.append((cap_j1, y))
    # 2. Encher completamente a jarra 2
    estados.append((x, cap_j2))
    # 3. Esvaziar completamente a jarra 1
    estados.append((0, y))
    # 4. Esvaziar completamente a jarra 2
    estados.append((x, 0))
    # 5. Transferir da jarra 1 para a 2
    transferir = min(x, cap_j2 - y)
    estados.append((x - transferir, y + transferir))
    # 6. Transferir da jarra 2 para a 1
    transferir = min(y, cap_j1 - x)
    estados.append((x + transferir, y - transferir))

    return estados

def bfs(cap_j1, cap_j2, objetivo):
    """Implementa a busca em largura (BFS) para encontrar a solução."""
    estado_inicial = (0, 0)

    # Verifica uma condição óbvia de "sem solução"
    if objetivo > max(cap_j1, cap_j2):
        print("\nNenhuma solução encontrada.")
        print(f"O objetivo ({objetivo}L) é maior que a capacidade da maior jarra ({max(cap_j1, cap_j2)}L).")
        return

    fila = deque()         # fila para armazenar os estados a serem explorados
    visitado = set()       # conjunto para armazenar os estados já visitados
    predecessor = {}       # dicionário para rastrear o caminho de cada estado

    fila.append(estado_inicial)
    visitado.add(estado_inicial)
    predecessor[estado_inicial] = None

    while fila:
        atual = fila.popleft()
        x, y = atual
        # print(f"Explorando: Jarra {cap_j1}L: {x}L, Jarra {cap_j2}L: {y}L")

        # Verifica se atingimos o objetivo em qualquer uma das jarras
        if x == objetivo or y == objetivo:
            print("\nSolução encontrada!")
            caminho = []
            while atual is not None:
                caminho.append(atual)
                atual = predecessor[atual]
            caminho.reverse()

            print("--- Caminho da Solução ---")
            for i, estado in enumerate(caminho):
                print(f"Passo {i}: Jarra {cap_j1}L: {estado[0]}L, Jarra {cap_j2}L: {estado[1]}L")
            return

        for estado_seguinte in obter_estados_seguintes(atual, cap_j1, cap_j2):
            if estado_seguinte not in visitado:
                visitado.add(estado_seguinte)
                predecessor[estado_seguinte] = atual
                fila.append(estado_seguinte)

    print("\nNenhuma solução encontrada para esta combinação.")

# --- Início da Interação com o Usuário ---
if __name__ == "__main__":
    print("--- Problema dos Jarros de Água ---")

    while True:
        try:
            jarra1_cap = int(input("Digite a capacidade máxima da primeira jarra: "))
            jarra2_cap = int(input("Digite a capacidade máxima da segunda jarra: "))
            objetivo_val = int(input("Digite o volume objetivo a ser medido: "))

            if jarra1_cap <= 0 or jarra2_cap <= 0 or objetivo_val < 0:
                print("Por favor, insira apenas valores positivos para as capacidades e um objetivo não negativo.")
                continue

            break # Sai do loop se as entradas forem válidas
        except ValueError:
            print("Entrada inválida. Por favor, digite apenas números inteiros.")

    # Executa o algoritmo de busca com os valores fornecidos pelo usuário
    bfs(jarra1_cap, jarra2_cap, objetivo_val)
```
